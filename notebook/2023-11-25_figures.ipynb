# Custom Color Palette

```python
import colorsys

# get colors
colors = plt.rcParams['axes.prop_cycle'].by_key()['color']
# convert from html to hsv
hsv = [colorsys.rgb_to_hsv(*matplotlib.colors.to_rgb(c)) for c in colors]


# show the colors
plt.figure(figsize=(5, 1))
for i, c in enumerate(colors):
    plt.plot([i], [0], "o", ms=30, color=c)
plt.axis("off")
plt.show()

hsv = np.array(hsv) 
hsv[-2] / hsv[-1] 
```

```python
colors = [
    '#00A9E0', 
    '#FEA090', 
    "#9AA0A7", 
    "#077187", 
    '#0A9086', 
    "#3E5496", 
    "#E0607E", 
    "#8E2043",
]


SEEBLAU_HSV = colorsys.rgb_to_hsv(*matplotlib.colors.to_rgb(SEEBLAU))
# sort by hue
colors = list(sorted(colors, key=lambda c: (colorsys.rgb_to_hsv(*matplotlib.colors.to_rgb(c))[0] - SEEBLAU_HSV[0]) % 1))
# convert from html to rgb
colors = [matplotlib.colors.to_rgb(c) for c in colors]

# show the colors
plt.figure(figsize=(5, 1))
for i, c in enumerate(colors):
    plt.plot([i], [0], "o", ms=20, color=c)
# plt.axis("off")
plt.ylim(-1, 1)
plt.show()
```

```python
# convert seeblau to hsv

hsv = colorsys.rgb_to_hsv(*matplotlib.colors.to_rgb("#59C7EB"))
# create variations with 5 different hues
n = 5
new_colors = [colorsys.hsv_to_rgb((hsv[0] - i/n) % 1, hsv[1], hsv[2] - i * .1/n) for i in range(n)]

# add variations of every color
new_colors = [
    *new_colors,
    *[
        colorsys.hsv_to_rgb(
            # c[0] * 1.02,
            # c[1] *  1,
            # c[2] * .91
            c[0] - 1/(2*n),
            c[1] *  1,
            c[2] * .8
        ) for c in [colorsys.rgb_to_hsv(*col) for col in new_colors]
    ],
]

# new_colors = [new_colors[0], matplotlib.colors.to_rgb('#9AA0A7'), *new_colors[1:]]

# shuffle the colors
#new_colors = list(sorted(new_colors, key=lambda c: (colorsys.rgb_to_hsv(*c)[0] - SEEBLAU_HSV[0]) % 1))

# save as default colors
plt.rcParams['axes.prop_cycle'] = plt.cycler(color=new_colors)



plt.figure(figsize=(5, 1))
for i, c in enumerate(new_colors):
    plt.plot([i], [0], "o", ms=20, color=c)
# plt.axis("off")
plt.ylim(-1, 1)
plt.show()

pprint(new_colors)
```
